```python
# Days out of the UK = Starting from being in the UK..
#                          (Date I went in to a different country) - (Next time I went in to the UK)


results = pd.DataFrame(columns=df.columns)
i=0
index2=0 # First UK in
for index, row in df.iterrows():
    # index==index2 so once it has found one 'in UK' and it finds the next one below, it continues from that one
    if index >= index2 and row['Country'] == 'UK' and row['Type'] == 'in' and row['Date'] >= initial: # Find a date I'm already in the UK
        # append and concat do not work, must use loc (not iloc)
        #results.loc[i] = row
        for index1, row1 in df.iterrows():
            if index1 > index and row1['Country'] != 'UK' and row1['Type'] == 'in':
                results.loc[i] = df.ix[index1] # Find the next date I left the UK
                for index2, row2 in df.iterrows():
                    if index2 > index1 and row2['Country'] == 'UK' and row2['Type'] == 'in':
                        results.loc[i+1] = df.ix[index2] # Find the date after that when I returned to the UK
                        i+=2
                        break
                break
        
print results

daysoutofUK = results['Date'].diff()
print daysoutofUK
print daysoutofUK.sum()
```
